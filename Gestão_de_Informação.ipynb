```python
import pandas as pd

# Suponha que dados_aba2 seja seu DataFrame e dados_abaNOVO seja o DataFrame recém-criado

# Função para inserir valores distintos de 'Campus/Polo/Unidade/Online' para cada 'Curso (título do card)'
def inserir_valores_campus_por_curso(dados_abaNOVO, dados_aba2):
    # Criar um DataFrame vazio para armazenar os resultados
    resultado = pd.DataFrame(columns=dados_abaNOVO.columns)

    # Para cada 'Curso (título do card)' em dados_abaNOVO
    for curso in dados_abaNOVO['Curso (título do card)'].unique():
        # Copiar os dados correspondentes ao curso de dados_aba2 e remover a coluna 'Campus/Polo/Unidade/Online'
        #dados_curso = dados_aba2[dados_aba2['Curso (título do card)'] == curso].drop(columns=['Campus/Polo/Unidade/Online'])

        # Para cada valor único de 'Campus/Polo/Unidade/Online' para este curso
        for campus in dados_curso['Campus/Polo/Unidade/Online'].unique():
            # Adicionar uma nova linha com os dados do curso e o valor de 'Campus/Polo/Unidade/Online'
            nova_linha = dados_curso.iloc[0].copy()  # Copiar a primeira linha como um modelo
            nova_linha['Campus/Polo/Unidade/Online'] = campus
            resultado = resultado.append(nova_linha, ignore_index=True)

    return resultado

# Chamar a função e atribuir o resultado a um novo DataFrame
dados_abaNOVO_com_campus = inserir_valores_campus_por_curso(dados_abaNOVO, dados_aba2)

# Exibir os dados do novo DataFrame
print(dados_abaNOVO_com_campus)
```
